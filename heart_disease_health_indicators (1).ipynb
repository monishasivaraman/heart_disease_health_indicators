**IQR (Interquartile Range) method**

```python
numeric_hd = hd.select_dtypes(include=['number'])  # Select only numeric columns

Q1 = numeric_hd.quantile(0.25) # Q1(First quartile) represents the 25th percentile, meaning 25% of the data points are below it

Q3 = numeric_hd.quantile(0.75) # Q3(second quartile)represents the 75th percentile, meaning 75% of the data points are below it

IQR = Q3 - Q1 #The IQR is the range between Q1 and Q3  it's calculated as:  IQR = Q3 - Q1

#data point below the lower bound or above the upper bound is considered a potential outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = numeric_hd[(numeric_hd < lower_bound) | (numeric_hd > upper_bound)]

# Identify which values are outliers
outliers_values = outliers.stack() # Used After Stack
print("Potential Outliers:")
print(outliers_values)

# Identify which values are not outliers
non_outliers = numeric_hd[~((numeric_hd < lower_bound) | (numeric_hd > upper_bound))]
non_outliers_values = non_outliers.stack()
print("\nNon-Outliers:")
print(non_outliers_values)
```
